```python
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook
```

```python
img_train = "./files/_temp_train_plot2.png"
img_prune = "./files/_temp_prune_plot2.png"
dpi = 100
im_data = plt.imread(img_train)
# img0, img1 = im_data, im_data
height, width, depth = im_data.shape
# What size does the figure need to be in inches to fit the image?
figsize = width / float(dpi), height / float(dpi) * 2 ## 2 images

fig = plt.figure(figsize=figsize)
ax2 = fig.add_axes([0,0,1,0.5])
ax = fig.add_axes([0,0.5,1,0.5])

# ax = fig.add_subplot(211)
# ax2 = fig.add_subplot(212)
plt.subplots_adjust(wspace=0, hspace=0)
plt.axis('off')

while True:
    try:
        img0 = plt.imread(img_train)
        ax.clear()
        ax.axis("off")
    #     ax.axis("equal")
        ax.imshow(img0)
    except:
        pass
    
    try:
        img1 = plt.imread(img_prune)
        ax2.clear()
        ax2.axis("off")
    #     ax2.axis("equal")
        ax2.imshow(img1)
    except:
        pass
    
    fig.canvas.draw()
    plt.pause(0.5)
```
