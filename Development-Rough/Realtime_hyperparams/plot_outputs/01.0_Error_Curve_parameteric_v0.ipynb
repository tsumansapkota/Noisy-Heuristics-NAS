```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def error_curve(x, a=0):
    return (1-x)*np.exp(a*x)
```

```python
xs = np.linspace(0, 1, 100)
```

```python
params = [1, 0.5, 0, -1, -2, -5, -10]
for a in params:
    ys = error_curve(xs, a)
    plt.plot(xs, ys, label=f"a={a}")
plt.xlabel("Normalized Epochs")
plt.ylabel("Normalized Loss")
plt.legend()
plt.savefig("./loss_curve.pdf")
plt.show()
```
